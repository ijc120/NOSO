### define the data frame function, it convert association files into pandas data frame

```python
def ar_df(association_rules):
    association_rules=list(association_rules)
    ar={}
    ar['item_base']=[]
    ar['item_add']=[]
    ar['support']=[]
    ar['confidence']=[]
    ar['lift']=[]

    for item in association_rules:
        item_base=[x for x in item[2][0][0]]
        ar['item_base'].append(item_base)

        item_add=[x for x in item[2][0][1]]
        ar['item_add'].append(item_add)

        support=item[1]
        ar['support'].append(support)

        confidence=item[2][0][2]
        ar['confidence'].append(confidence)

        lift=item[2][0][3]
        ar['lift'].append(lift)

    ar=pd.DataFrame(ar)

    return ar
```

### how to perform recommandation
1. create an order list
2. for each item in the list, create an list containing its associated items (the highest support)
3. check if there is sale (discount) among these associated items
   a) if yes, recommand sale items
   b) if no, return the highest support item

```python
#create sale list, randomly select 100 items as sale item
sale=usjh_all[['lineitem_name']].sample(100)
sale.head()
```

```python
#create sale vendor: category based on sale items
sale_vc=[]
for item in sale['lineitem_name']:
    sale_vc.append(usjh_all[usjh_all['lineitem_name']==item].vendor_category.iloc[0])
    
sale_vc=set(sale_vc)
```
